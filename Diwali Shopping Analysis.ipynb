```python
import pandas as pd #work with dataframes
import numpy as np #work with numericals
import matplotlib.pyplot as plt #visualozation
%matplotlib inline
import seaborn as sns #visualization
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
#to avoid encoding error use 'unicode_escape'
```

```python
df.shape #shows no.of rows,columns
```

```python
df.head()
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
#axis=1: full row, inplace=True: save the changes
```

```python
df.info()
```

```python
pd.isnull(df)#to check null values in data
```

```python
pd.isnull(df).sum()#to find no. of null values in each coulmn
```

```python
df.shape #no. of rows before removing null values
```

```python
df.dropna(inplace=True)# drop missing values
```

```python
df.shape #no. of rows after removing null values
```

```python
pd.isnull(df).sum()
```

```python
#change datatype
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#remane coulumn
df.rename(columns={'Marital_Status':'Shaadi'}) #dictionary in key:value pair
```

```python
df.describe()#gives statistical values for all the columns containing numericals
```

```python
#use describe for specific columns
df[['Age', 'Orders']].describe()
```

```python
ax = sns.countplot(x='Gender', data=df)#data from df
sns.set(rc={'figure.figsize':(2,4)})

for bars in ax.containers:
    ax.bar_label(bars)
    #gives the nos on the bars
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc={'figure.figsize':(5,5)})

sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
ax=sns.countplot(x='Age Group', hue='Gender', data=df, gap=.2)

for bars in ax.containers:
    ax.bar_label(bars)
   
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
#top 10 states based on orders placed
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending= False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Orders', data=sales_state)
```

```python
#top 10 states based on amount spent
sales_state_amt = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Amount', data=sales_state_amt)
```

```python
ax = sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(2,3)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=sales_marital_status)
```
